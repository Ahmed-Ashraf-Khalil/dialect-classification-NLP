# Importing the tools

```python
import numpy as np
import pandas as pd
import plotly.express as ex
import requests
import json
```

# Read the Dataframe given

```python
df = pd.read_csv("dialect_dataset.csv")
```

```python
#change the type of it
df["id"] = df.id.astype("str")
```

# Data fetish and mapping

```python
# containers
l = {}
d=0

#data gathering
for i in range(1,460):
    try:
        data = json.dumps(list(df.iloc[0+d:1000+d,0]))
        post = requests.post("https://recruitment.aimtechnologies.co/ai-tasks",data=data)
        l.update(post.json())        
        d+=1000
        print(len(l),": Exampled got from the API")
    except:
        data = json.dumps(list(df.iloc[1000+d:,0]))
        post = requests.post("https://recruitment.aimtechnologies.co/ai-tasks",data=data)
        l.update(post.json())
        print(len(l),": Exampled got from the API")
        
df["Comment"] = df["id"].map(l)
```
